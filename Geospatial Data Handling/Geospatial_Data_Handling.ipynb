## **Visualizing Level-1 Administrative Divisions in Israel**

In this section, we will load and visualize the level-1 administrative divisions of Israel using a shapefile from the GADM database. The GADM (Global Administrative Areas) database provides high-resolution maps of countries and their respective administrative areas. The level-1 administrative divisions typically represent regions or provinces within a country.

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx
```

A shapefile is a popular geospatial vector data format for geographic information system (GIS) software. A shapefile is actually a collection of several files with the same base name but different extensions. The essential components of a shapefile are:





1.  **.shp (Shape format):-** This file contains the geometry data for all the features (points, lines, or polygons).
2.   **.shx (Shape index format):-** This file contains the index of the geometry data.
3. **.dbf (Attribute format):-** This file contains attribute data for the features, stored in tabular format.
4. **.prj (Projection format):-** This file contains the coordinate system and projection information.
5. **.cpg (Code page format):-** This file specifies the character encoding used for the attribute data in the .dbf file.

```python
shapefile = '/content/gadm41_ISR_1.shp'
```

```python
gdf = gpd.read_file(shapefile)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
gdf.plot(ax=ax, color='orange', edgecolor='black', alpha=0.5)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
gdf.plot(ax=ax, color='lightblue', edgecolor='black', alpha=0.5)
ctx.add_basemap(ax, crs=gdf.crs.to_string(), source=ctx.providers.OpenStreetMap.Mapnik)
ax.set_title('Level-1 Administrative Divisions in Israel')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

plt.show()
```
